```python
import re

def extract_text_from_md(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        content = file.readlines()
    
    extracted_text = []
    paragraph = []
    
    for line in content:
        line = line.strip()
        
        # Ignore Markdown-specific elements and figure/table descriptions
        if not line or line.startswith(('#', '-', '*', '>', '|', '`', '[', '<', '!')) or re.match(r'^(Figure|Table) \d+ \|', line) or re.match(r'^\d+\.', line):
            if paragraph:
                extracted_text.append(" ".join(paragraph))
                paragraph = []
            continue
        
        # Remove inline HTML tags and superscripts
        line = re.sub(r'<.*?>', '', line)  # Remove HTML tags
        line = re.sub(r'\[.*?\]\(.*?\)', '', line)  # Remove inline links
        line = re.sub(r'\<sup>.*?</sup>', '', line)  # Remove superscripts
        
        # Collect sequential text parts
        paragraph.append(line)
    
    if paragraph:
        extracted_text.append(" ".join(paragraph))
    
    return "\n\n".join(extracted_text)

# Example usage
if __name__ == "__main__":
    file_path = "/kaggle/input/sourcesmd/mds/2501.12948v1.md"  # Change to your file path
    text_content = extract_text_from_md(file_path)
    print(text_content)
```
